# Convert Wav files

```python
import os
import librosa
import soundfile as sf

TARGET_SR = 16000

ROOT_DIR = "dataset"  

def process_wav(input_path):
    audio, sr = librosa.load(
        input_path,
        sr=TARGET_SR,
        mono=True
    )

    sf.write(
        input_path,
        audio,
        TARGET_SR,
        subtype="PCM_16"
    )

def process_folder(root_dir):
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.lower().endswith(".wav"):
                full_path = os.path.join(root, file)
                try:
                    process_wav(full_path)
                    print(f"✔ Processed: {full_path}")
                except Exception as e:
                    print(f"✖ Failed: {full_path} | {e}")

process_folder(ROOT_DIR)
```
